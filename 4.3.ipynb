```python
import numpy as np
from sklearn.mixture import GaussianMixture
import pickle
import matplotlib.pyplot as plt
```

```python
# generated from ChatGPT


def generate_sample_dataset(n, means, covariances, random_seed=None):
    if random_seed is not None:
        np.random.seed(random_seed)

    sample_data = []
    for i in range(n):
        component_data = np.random.multivariate_normal(means[i], covariances[i], size=100)
        sample_data.append(component_data)

    return np.vstack(sample_data)

data = None
with open ('./SMAI-Dataset-problem-4.3/colors.pkl', 'rb') as f:
    data = pickle.load(f)

n_components = 7

gmm = GaussianMixture(n_components=n_components)
gmm.fit(data)
means = gmm.means_
covariances = gmm.covariances_
labels = gmm.fit_predict(data)

print(means)
print(covariances)
print(labels)

plt.figure(figsize=(12, 6))

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')


sample_dataset = generate_sample_dataset(n_components, means, covariances)
sample_dataset
```
